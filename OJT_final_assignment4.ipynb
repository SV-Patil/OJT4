### Q4. Polynomial plotting with annotations.
Randomly select coefficients of 5th degree polynomial. set the random seed as last two digit of roll number. Plot polynomial for -5<= x <= 5. Annotate to plot to identify the maxima and minima of the polynomial.

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
np.random.seed(16)

coff = np.random.randint(1,9,6)

X = np.arange(-5,5+0.1,0.1)
Y = coff[0]*(X**5) + coff[1]*(X**4) + coff[2]*(X**3) + coff[3]*(X**2) + coff[4]*(X) +coff[5]

plt.plot(X,Y, color='blue')
plt.xlabel('X')
plt.ylabel('Y')
plt.annotate("Minima", xy=(-5,-2000), xytext=(-4,4000),arrowprops = dict(arrowstyle="->",facecolor ='blue'))
plt.annotate("Maxima", xy=(4.9,11000), xytext=(3,9000),arrowprops = dict(arrowstyle="->",facecolor ='red'))

plt.show()
```
